```python
base_model = HistGradientBoostingRegressor(random_state=123, max_depth=7, max_iter=100, learning_rate=0.1)

# Envolver el modelo base dentro de MultiOutputRegressor
multi_output_model = MultiOutputRegressor(base_model)
```
