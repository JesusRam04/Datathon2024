# Feature engineering

```python
flights['STD'] = pd.to_datetime(flights['STD'])
flights['STA'] = pd.to_datetime(flights['STA'])
```

```python
year_to_export = 2024
```

```python
#flights_2023 = flights[(flights['STA'].dt.year == year_to_export)]
```

```python
flights_2023 = flights[(flights['STA'].dt.year == 2024) & (flights['STA'].dt.month == 1)]
```

```python
flights_2023 = flights_2023.dropna(subset=['Origin_Type'])
flights_2023 = flights_2023.dropna(subset=['Destination_Type'])
```

```python
productos_deseados = ['Botanas', 'Licores', 'Galletas', 'Bebidas Calientes', 'Perecederos', 'Refrescos', 'Sopas', 'Lacteos']
sales_clean = sales[sales['ProductType'].isin(productos_deseados)]
sales_clean
```

```python
id_sales = sales_clean['Flight_ID'].unique()
len(id_sales)
```

```python
no_en_sales = ~flights_2023['Flight_ID'].isin(id_sales)

flights_no_en_sales = flights_2023.loc[no_en_sales, 'Flight_ID'].unique()

print(flights_no_en_sales)
len(flights_no_en_sales)
```

```python
vuelos_no_sales = flights_2023[flights_2023['Flight_ID'].isin(flights_no_en_sales)]

vuelos_no_sales
```

```python
vuelos_no_sales.reset_index(inplace=True)
```

```python
products_name = sales['ProductName'].unique()
```

```python

sales_pivot = sales_clean.pivot_table(index='Flight_ID', columns='ProductName', values='Quantity', fill_value=0)

merged_df = pd.merge(flights_2023, sales_pivot, on='Flight_ID', how='left')

merged_df
```

```python
columnas_no_reemplazar = ['Flight_ID', 'Aeronave', 'DepartureStation', 'ArrivalStation', 'Destination_Type', 'Origin_Type', 'STD', 'STA', 'Capacity', 'Passengers']

data_clean = merged_df.fillna({col: 0 for col in merged_df.columns if col not in columnas_no_reemplazar})
data_clean
```

```python
from sklearn.preprocessing import LabelEncoder

# Supongamos que tienes un DataFrame llamado flights_2023 con una columna 'Origin_Type'
# y deseas codificarla utilizando LabelEncoder

# Crear una instancia del codificador
le = LabelEncoder()

# Ajustar el codificador a los datos de la columna 'Origin_Type'
le.fit(data_clean['Destination_Type'])

# Transformar los valores categóricos en valores numéricos
data_clean['Destination_type_encoded'] = le.transform(data_clean['Destination_Type'])
del[data_clean['Destination_Type']]

# Ahora flights_2023 contiene una nueva columna 'Origin_type_encoded' con los valores codificados
```

```python
import holidays
```

```python
mx_holidays = holidays.Mexico()

# Función para marcar si una fecha es un día festivo en México
def is_mexico_holiday(fecha):
    if fecha in mx_holidays:
        return 1
    else:
        return 0

# Aplicar la función a cada fila de la columna 'fecha' y almacenar el resultado en una nueva columna 'is_holiday'
data_clean['is_holiday'] = data_clean['STD'].apply(is_mexico_holiday)
```

```python
data_clean['hour_STD'] = data_clean['STD'].dt.hour
data_clean['minutes_STD'] = data_clean['STD'].dt.minute
data_clean['hour_STA'] = data_clean['STA'].dt.hour
data_clean['minutes_STA'] = data_clean['STA'].dt.minute

data_clean['day_of_week'] = data_clean['STD'].dt.dayofweek  # 0 para lunes, 1 para martes, ..., 6 para domingo
data_clean['week_of_year'] = data_clean['STD'].dt.isocalendar().week
data_clean['month'] = data_clean['STD'].dt.month

data_clean['time_of_fligth'] = (data_clean['STA'] - data_clean['STD']).dt.total_seconds() / 60
```
